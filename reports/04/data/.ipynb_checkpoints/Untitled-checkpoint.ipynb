```python
import pandas as pd
import numpy as np
df = pd.read_csv('measurements_dev2.csv', sep=';', skiprows = 1, names=['platform', 'device', 'work_x', 'work_y', 'ex1', 'ov1','ex2', 'ov2','ex3', 'ov3','ex4', 'ov4', 'ex5', 'ov5'])
df.
```

```python
data = df[['work_x', 'work_y']]
data['ex'] =  df[['ex1', 'ex2', 'ex3', 'ex4', 'ex5']].mean(axis=1)
data
```

```python
data = data.pivot('work_x', 'work_y', 'ex')
data
```

```python
import seaborn as sns
sns.set()
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
ax = sns.heatmap(data, cmap=cmap)
ax.invert_yaxis()
```
