```python
import pandas as pd
import numpy as np
df = pd.read_csv('measurements-local.csv', sep=';', skiprows = 1, names=['platform', 'device', 'work_x', 'work_y', 'ex1', 'ov1','ex2', 'ov2','ex3', 'ov3','ex4', 'ov4', 'ex5', 'ov5'])
```

```python
data = df[['work_x', 'work_y']]
data['ex'] =  df[['ex1', 'ex2', 'ex3', 'ex4', 'ex5']].mean(axis=1)
data
```

```python
data = data.drop_duplicates(subset=['work_x', 'work_y'])
data
data = data.pivot('work_x', 'work_y', 'ex')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(9,7)) # does not affect the following plot

sns.set()
cmap = sns.cubehelix_palette(3, light=1, as_cmap=True)
ax = sns.heatmap(data, cmap=cmap, )
ax.invert_yaxis()
fig.savefig('output_powers.pdf')
```
